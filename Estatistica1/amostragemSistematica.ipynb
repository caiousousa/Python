```python
import numpy as np
import pandas as pd
from math import ceil
```

```python
populacao = 150
amostra=15
k=ceil(populacao/amostra)
print(k)
```

```python
r = np.random.randint(low=1, high=k+1,size=1)
print(r)
```

```python
acumulador = r[0]
sorteados=[]


for i in range(amostra):
    sorteados.append(acumulador)
    acumulador = acumulador + k
print(sorteados)
```
